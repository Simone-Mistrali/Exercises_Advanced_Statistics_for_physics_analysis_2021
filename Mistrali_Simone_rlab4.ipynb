## Ex1


Create Boxes and probabilities for extracting `W` or `B` in the boxes.

```r
H0 <- rep(c("B"),c(5))
H1 <- rep(c('B','W'), c(4,1)) 
H2 <- rep(c('B','W'), c(3,2)) 
H3 <- rep(c('B','W'), c(2,3))
H4 <- rep(c('B','W'), c(1,4))
H5 <- rep(c("W"),c(5))
probW <- c(0,1,2,3,4,5)*1/5
```

```r
bayes_th <- function(prob, extr) {
    if (extr == 'W') {
        num <- probW * prob
        den <- sum(probW * prob)
        }
    else {
        num <- (1-probW) * prob
        den <- sum((1-probW) * prob)
    }
    return(num/den)
}
f <- function(c) {
    prob <- rep(1/6,6)
    ex <- prob
    for (i in 1:c) {
        extrac <- readline(prompt="What color have you drawn? ")
        prob <- bayes_th(prob, extrac)
        ex <- append(ex, prob)
        }
    ex <- matrix(ex,nrow=as.integer(c)+1,ncol=6,byrow=TRUE)
    return (ex) 
}
c <- readline(prompt="How many extraction? ")
ex <- f(c)
```

## Ex 3

Given a sequence of indipendet random variable $X_j \sim Gamma(\alpha_j,\beta)$ because of the _reproductive property_ of the Gamma distribution it follows that
$$
Y=\sum_{j=1}^n X_j\, \rightarrow\, Y\sim Gamma(\alpha,\beta)\, \text{ where }\, \alpha=\sum_{j=1}^n=\alpha_j
$$

If $\alpha=m \in \mathbb{N}$, a random variable from gamma distribution, in this case called Erland distribution, can be obtained by summing up $m$ independint exponential random variables $X_j\sim Exp(\beta)$:
$$
Y=\beta \sum_{j=1}^n(-\ln(U_j) = -\beta \ln \prod_{j=1}^nU_j
$$
where $U_j$ is a random variable generated by a uniform distribution.

```r
erlang <- function(n,m=30,beta=1){
               Y <- NULL
               for(i in 1:n){
                   x <- runif(m)
                   x.log <- -log(x)
                   Y.val <- beta*sum(x.log)
                   Y <- c(Y,Y.val) }
               return(Y)
               }
```

```r
n <- 100000
m <- 100
beta <- 6

x1 <- erlang(n,m=m,beta=beta)
# If the algorithm is correct x1 and x2 need to be similar
x2 <- rgamma(n,shape=m,rate=beta)


options(repr.plot.width=16, repr.plot.height=10)

hist(x1, col="#7209B7", xlab='x', ylab='Density', density=100, 
        main="Sampling from Erlang distribution", prob=TRUE)

hist(x2, col="#FF0035", xlab='x', ylab='Density', density=70,
        main="Sampling from Erlang distribution", prob=TRUE, add=TRUE)

legend("topright", inset=0.1,legend=c("Erlang","Gamma"),
       col=c("#7209B7","#FF0035"), cex=0.9,
       fill=c("#7209B7","#FF0035"), box.lty=1)

grid()
box()
```

## Ex 4

In mathematics, the **middle-square method** is a method of generating pseudorandom numbers. This method work as follows, a number (`seed`) is squared and in the middle of the number is taken a number of the same lenght of the previous, e.g.:


$$
        seed=675248\\
        (seed)^2 = 455\underbrace{959861}_{\text{new seed}}504\\
        new\_seed=959861
$$

```r
MSM <- function(x,counter=1,verbose=FALSE){

    for (i in 1:counter){
        len <- as.integer(length(unlist(strsplit(sprintf("%d",x),""))))
        new_seed <- unlist(strsplit(as.character(x^2),""))
        lenn <- length(new_seed)%/%2
        
        if (len %% 2 ==0){
        new_seed <- new_seed[(lenn-len%/%2+1) : (lenn+len%/%2)]
    
        }
        else{
            new_seed <- new_seed[(lenn-len%/%2+1) : (lenn+len%/%2+1)]
        }
        new_seed<- as.numeric(paste(new_seed,collapse=""))
        if (verbose==TRUE) print(paste0("The seed at ",i," iteration is ",new_seed))
        x<- new_seed
        
    }
    return (new_seed)
}
```
